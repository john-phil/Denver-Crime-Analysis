```python
# Function to find number of missing values in a column
def count_missing_values(neighborhood_id):
    return crime[neighborhood_id].isna().sum()

# Example usage
missing_values_column1 = count_missing_values('neighborhood_id')

print(f"Number of missing values in Column1: {missing_values_column1}")
```

```python
crime_cleaned=crime_clean[crime_clean['geo_lon'].isnull()==False]
```

```python
demo=pd.read_csv(os.path.join(path, 'Data', 'Original Data', 'Denver_Demo_Data.csv'), index_col=False)
```

```python
demo.head()
```

```python
demo.rename(columns= {'Neighborhood ID': 'neighborhood_id'}, inplace=True)
```

```python
#merging denver crime dataset with denver demographics datset using the neighborhood id as the key column.
```

```python
crime_merged=crime_cleaned.merge(demo, on= 'neighborhood_id')
```

```python
crime_merged.head()
```

```python
pd.set_option('display.max_columns', None)
```

```python
crime_merged.describe()
```

```python
crimes_merged=crime_merged.drop(columns= ['geo_x', 'geo_y'])
```

```python
crimes_merged.head()
```

```python
# Group by 'Neighborhood Name' and calculate the required statistics
agg_df = crimes_merged.groupby('Neighborhood Name').agg({
    'victim_count': ['mean', 'min', 'max', 'std', 'count']
})

# Sort by mean victim_count in descending order
sorted_df = agg_df.sort_values(('victim_count', 'mean'), ascending=False)

# Set display option to show all rows
pd.set_option('display.max_rows', None)

print(sorted_df)
```

```python
sorted_df.to_clipboard()
```

```python
#this shows the neighborhoods ranked by highest mean number of victims per crime
```

```python
#finding the top 10 neighborhoods by number of crimes
df_agg_2= crimes_merged.groupby('Neighborhood Name').agg({'Neighborhood Name': ['count']})
sorted_agg=df_agg_2.sort_values(('Neighborhood Name', 'count'), ascending=False)
top_10_df=sorted_agg.head(10)
top_10_df
```

```python
top_10_df.to_clipboard()
```

```python
#finding the bottom 10 neighborhoods by number of crimes
bottom_10_df=sorted_agg.tail(10)
bottom_10_df
```

```python
bottom_10_df.to_clipboard()
```
